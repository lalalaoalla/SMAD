```python
import pandas as pd
import numpy as np

df=pd.read_csv('lab_1_data.csv')
df
```

```python
#предлагаю если что оставшийся импорт библиотек сделать здесь
import scipy.stats
```

```python
#наша новая функция(с доп регрессором)
def u_func(x1, x2):
    return 2 + 2.5*x1 + 0.03*x1*x1 + 0.1*x2 - 2*x2*x2+0.001*x2**3
```

```python
#наши старые х1 и х2
x1 = np.array([-1, -0.5, 0.5, 1])
x2 = np.array([-1, -0.5, 0, 0.5, 1])
u = np.array([])

for i in x1:
    for j in x2:
        u = np.append(u, u_func(i, j))
```

```python
#изменила генерацию e, теперь в функцию подается среднеквадратичное отклонение
n = len(u)
p = 0.1 # доля
u_average = np.full(n, np.mean(u))

# считаем ошибку наблюдения
w_sq = np.dot((u - u_average).transpose(), (u - u_average)) / (n - 1)
d = (p * w_sq)**(1/2)
e = np.random.normal(0, d, n)
```

```python
# дополнительные столбцы в датафрейме
df['x1^2'] = df['x1']**2
df['x2^2'] = df['x2']**2
df['x2^3'] = df['x2']**3
df
```

```python
class MyLinearRegression:
    
    def __init__(self):
        self.coef_ = None
        self.intercept_ = None
        self.k = None
        self.for_sigma = None
        
        
    def fit(self, X, y):
        X = np.array(X)
        y = np.array(y)

        X = np.hstack((np.ones((X.shape[0], 1)), X))

        self.k = np.linalg.inv(X.T @ X) @ X.T @ y
        self.for_sigma = np.linalg.inv(X.T @ X)

        self.coef_ = self.k[::][1:]
        self.intercept_ = self.k[::][0]
        
    def predict(self, X):
        X = np.hstack((np.ones((X.shape[0], 1)), X))
        y_pred = X @ self.k        
        return y_pred
    
    def get_param(self):
        return self.coef_, self.intercept_
```

```python
LinModel = MyLinearRegression()
```

```python
#добавила в х новый регрессор
x = df[['x1','x1^2','x2', 'x2^2','x2^3']]
y = df['y']
```

```python
LinModel.fit(x,y)
y_predict = LinModel.predict(x)
```

```python
LinModel.get_param()
```

```python
tetta_wave_j = LinModel.get_param()[1] 
tetta = LinModel.get_param()[0] 
tetta = np.append(tetta_wave_j, tetta) 
print(tetta)
```

### Для вычисления сигмочек и тд

```python

sigma = np.sqrt(p*w_sq)
e_predict = y - y_predict
sigma_sq_predict = e_predict.T @ e_predict / (n - 6)
print(sigma_sq_predict)
F = sigma_sq_predict/sigma**2
FT = scipy.stats.f.ppf(q=1-0.05, dfn=9999, dfd=16)
print('F =', F)
print('FT =', FT)
print('F < FT:', F < FT)
```

## Четвертый пункт. Проверка гипотезы о незначимости самой регрессии

```python
# theta - массив найденных параметров
theta_intercept = LinModel.get_param()[1]
theta = LinModel.get_param()[0]
theta = np.append(theta_intercept, theta)
theta
```

```python
# Формулы:
# F = (RSSh - RSS) * (n - m) / q / RSS
# RSSh = sum( ( yi - y_aver )^2 )
# RSS = (y - X*theta)T * (y - X*theta)

m = theta.size
q = m - 1
X_wout_intercept = np.array(x)
X_arr = np.hstack((np.ones((X_wout_intercept.shape[0], 1)), X_wout_intercept))

rssh = 0
rss = np.dot((y - np.dot(X_arr, theta)).T, (y - np.dot(X_arr, theta)))

# y_aver - среднее истинного y
y_aver = np.average(y)
for i in range(n):
  rssh += pow((y[i] - y_aver), 2)

F_regr_level = ((rssh - rss) / q) / (rss / (n - 6))
FT_regr_level = scipy.stats.f.ppf(q=1-0.05, dfn=q, dfd=n-m)

print('Незначимость регресии: ')
print('F = ', F_regr_level)
print('FT = ', FT_regr_level)
print('F < FT', F_regr_level < FT_regr_level)
```

## Пятый и шестой пункты. Доверительное оценивание для мат. ожидания
#### для истинного значения математического ожидания функции отклика в точке фиксированной

```python
nuMin = []
nuOrig = []
nuMax = []
```

```python
# фиксируем x1 = -1
X1 = x1[0]
X2 = x2[0]

# оригинальная тетта
theta_original = np.array([2, 2.5, 0.03, 0.1, -2, 0.001])

# Считаем корень из F(alpha, 1, fR)
F_di = pow(scipy.stats.f.ppf(q=1-0.05, dfn=1, dfd=n-m), 0.5)

i = 0
while (X2 <= x2[4]):
    func = np.array([1, X1, X1**2, X2, X2**2, X1**3])
    # считаем сигму
    sigma_new = pow(sigma_sq_predict, 0.5) * pow(np.dot(np.dot(func.T, np.linalg.inv(np.dot(X_arr.T, X_arr)) ), func), 0.5) 
    
    # считаем этту
    nu = np.dot(func.T, theta)

    # считаем границы
    nu_min = nu - F_di * sigma_new
    nu_max = nu + F_di * sigma_new
    # что есть этта оригинал?
    nu_original = np.dot(func.T, theta_original)
    X2 = X2 + 0.2

    # чтобы удобней переносить в эксель было
    print(f'{nu_min},{nu_original},{nu_max}')
    #Для графиков
    nuMin.append(nu_min)
    nuOrig.append(nu_original)
    nuMax.append(nu_max)
    i += 1
```
